```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
#Scatter Plot
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
#Line Plot
sns.lineplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
# Histogram
sns.histplot(tips["total_bill"], bins=20)
plt.show()
```

```python
# Bar Plot
sns.barplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
#Box Plot
sns.boxplot(data=tips, x="day", y="total_bill")
plt.show()
```

```python
#Pair Plot
sns.pairplot(tips)
plt.show()
```

```python
sns.set_style("whitegrid")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Scatter Plot of Total Bill vs Tip")
plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.show()
```

```python
sns.set_palette("pastel")
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.show()
```

```python
#Heatmap
corr = tips.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```

```python
#Facet Grid
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
plt.show()
```

```python
#Saving Plots
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.savefig("scatterplot.png")
```

```python
# Create a joint plot
sns.jointplot(data=tips, x="total_bill", y="tip", kind="hex", color="k")
plt.show()
```

```python
#Pair Plot with Hue
sns.pairplot(tips, hue="sex", palette="coolwarm")
plt.show()
```

```python
#Violin Plot
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.show()
```

```python
#Facet Grid
g = sns.FacetGrid(tips, col="time", row="smoker")
g.map(sns.histplot, "total_bill")
plt.show()
```

```python
#Swarm Plot
sns.swarmplot(data=tips, x="day", y="total_bill", hue="sex", dodge=True)
plt.show()
```

```python
#Heatmap with Annotations
corr = tips.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, annot=True, cmap="coolwarm", linewidths=.5)
plt.show()
```

```python
#Regression Plots (with Multiple Levels)
sns.lmplot(data=tips, x="total_bill", y="tip", hue="smoker", markers=["o", "x"], palette="coolwarm", height=7)
plt.show()
```

```python
#Cat Plot
sns.catplot(data=tips, x="day", y="total_bill", hue="sex", kind="point", col="time", dodge=True)
plt.show()
```

```python
#Time Series Plot

# Load example data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(data=fmri, x="timepoint", y="signal", hue="region", style="event")
plt.show()
```

```python
#Customizing Figure Aesthetics
# Set the context to "talk" for larger text
sns.set_context("talk")

# Set the style to "ticks"
sns.set_style("ticks")

# Set a custom color palette
sns.set_palette("pastel")

# Create a violin plot with the new settings
sns.violinplot(data=tips, x="day", y="total_bill", hue="sex", split=True)
plt.show()
```
